## 배분 시각화

```python
# visualizer.py 방식으로 data_real.csv 시각화 (각 PROD_CD별로 따로)
print("📊 data_real.csv 시각화 시작...")

import os
import matplotlib.pyplot as plt
import numpy as np

# output 폴더 생성
output_dir = "output"
os.makedirs(output_dir, exist_ok=True)

# 1. SKU 컬럼 생성 (PROD_CD_COLOR_CD_SIZ_CD)
df['SKU'] = df['PROD_CD'] + '_' + df['COLOR_CD'] + '_' + df['SIZ_CD']

# 2. 매장별 총 QTY 계산 (QSUM 역할)
store_qty_sum = df.groupby('TO_LGT_SHOP_CD')['QTY'].sum().to_dict()

print(f"총 매장 수: {len(store_qty_sum)}")
print(f"총 SKU 수: {df['SKU'].nunique()}")
print(f"총 PROD_CD 수: {df['PROD_CD'].nunique()}")

# 3. 각 PROD_CD별로 시각화 함수 정의
def create_allocation_matrix_heatmap_by_prod(df, prod_cd, output_dir):
    """
    특정 PROD_CD에 대한 배분 매트릭스 히트맵 생성
    """
    print(f"\n📊 {prod_cd} 배분 매트릭스 히트맵 생성 중...")
    
    # 해당 PROD_CD만 필터링
    df_prod = df[df['PROD_CD'] == prod_cd].copy()
    
    if len(df_prod) == 0:
        print(f"   ⚠️ {prod_cd}에 대한 데이터가 없습니다.")
        return None
    
    # 매장과 SKU 목록 추출
    target_stores = df_prod['TO_LGT_SHOP_CD'].unique().tolist()
    SKUs = df_prod['SKU'].unique().tolist()
    
    # 매장별 QTY 합계 (해당 PROD_CD만)
    store_qty_sum_prod = df_prod.groupby('TO_LGT_SHOP_CD')['QTY'].sum().to_dict()
    
    # 배분 매트릭스 생성
    allocation_matrix = {}
    for _, row in df_prod.iterrows():
        sku = row['SKU']
        store = row['TO_LGT_SHOP_CD']
        qty = row['QTY']
        allocation_matrix[(sku, store)] = qty
    
    # 1. 매장들을 QTY_SUM 기준으로 정렬
    allocated_stores = []
    for store in target_stores:
        store_total = sum(allocation_matrix.get((sku, store), 0) for sku in SKUs)
        if store_total > 0:
            allocated_stores.append((store, store_total, store_qty_sum_prod[store]))
    
    allocated_stores.sort(key=lambda x: x[2], reverse=True)
    selected_stores = [store[0] for store in allocated_stores]
    
    # 2. SKU들을 컬러-사이즈 기준으로 정렬
    allocated_skus = []
    for sku in SKUs:
        sku_total = sum(allocation_matrix.get((sku, store), 0) for store in selected_stores)
        if sku_total > 0:
            parts = sku.split('_')
            color = parts[1] if len(parts) >= 3 else 'Unknown'
            size = parts[2] if len(parts) >= 3 else 'Unknown'
            allocated_skus.append((sku, sku_total, color, size))
    
    def get_size_sort_key(size):
        text_sizes = {'XS': 1, 'S': 2, 'M': 3, 'L': 4, 'XL': 5, 'XXL': 6}
        if size in text_sizes:
            return (0, text_sizes[size])
        try:
            numeric_size = int(size)
            return (1, numeric_size)
        except:
            return (2, size)
    
    allocated_skus.sort(key=lambda x: (x[2], get_size_sort_key(x[3])))
    selected_skus = [sku[0] for sku in allocated_skus]
    
    # 3. 매트릭스 데이터 생성
    matrix_data = []
    store_labels = []
    for store in selected_stores:
        row = [allocation_matrix.get((sku, store), 0) for sku in selected_skus]
        matrix_data.append(row)
        store_labels.append(f"{store}\\n({store_qty_sum_prod[store]:,})")
    
    # 4. SKU 라벨 생성
    sku_labels = []
    for sku in selected_skus:
        parts = sku.split('_')
        color = parts[1] if len(parts) >= 3 else 'Unknown'
        size = parts[2] if len(parts) >= 3 else 'Unknown'
        total_allocated = sum(allocation_matrix.get((sku, store), 0) for store in target_stores)
        sku_labels.append(f"{color}-{size}\\n({total_allocated})")
    
    # 5. 부가 통계 계산
    total_colors = df_prod['COLOR_CD'].nunique()
    total_sizes = df_prod['SIZ_CD'].nunique()
    
    empty_cells_counts = []
    color_cov_ratios = []
    size_cov_ratios = []
    
    for row_idx, store in enumerate(selected_stores):
        row_qties = matrix_data[row_idx]
        empty_cells_counts.append(row_qties.count(0))
        
        # 색상/사이즈 커버리지
        allocated_skus_row = [selected_skus[col_idx] for col_idx, qty in enumerate(row_qties) if qty > 0]
        colors = set()
        sizes = set()
        for sku in allocated_skus_row:
            parts = sku.split('_')
            if len(parts) >= 3:
                colors.add(parts[1])
                sizes.add(parts[2])
        
        color_cov_ratios.append(len(colors)/total_colors if total_colors else 0)
        size_cov_ratios.append(len(sizes)/total_sizes if total_sizes else 0)
    
    avg_empty_cells = np.mean(empty_cells_counts) if empty_cells_counts else 0
    avg_color_cov = np.mean(color_cov_ratios) if color_cov_ratios else 0
    avg_size_cov = np.mean(size_cov_ratios) if size_cov_ratios else 0
    
    # 6. 히트맵 생성
    matrix_data = np.array(matrix_data)
    vmax_val = max(1, matrix_data.max())
    
    # 한글 폰트 설정
    plt.rcParams['font.family'] = 'DejaVu Sans'
    plt.rcParams['axes.unicode_minus'] = False
    
    # 동적 크기 설정
    fig_width = max(12, len(selected_skus) * 0.8)
    fig_height = max(8, len(selected_stores) * 0.4)
    
    fig, ax = plt.subplots(figsize=(fig_width, fig_height))
    im = ax.imshow(matrix_data, cmap='Blues', aspect='auto', vmin=0, vmax=vmax_val)
    cbar = plt.colorbar(im, ax=ax, shrink=0.8)
    cbar.set_label('Allocated Quantity', rotation=270, labelpad=15)
    
    ax.set_xticks(range(len(selected_skus)))
    ax.set_yticks(range(len(selected_stores)))
    ax.set_xticklabels(sku_labels, rotation=45, ha='right', fontsize=9)
    ax.set_yticklabels(store_labels, ha='right', fontsize=9)
    
    # 셀에 숫자 표시
    for i in range(len(selected_stores)):
        for j in range(len(selected_skus)):
            qty = matrix_data[i, j]
            if qty > 0:
                text_color = 'white' if qty > matrix_data.max()*0.6 else 'black'
                ax.text(j, i, str(int(qty)), ha='center', va='center', 
                       color=text_color, fontweight='bold', fontsize=8)
    
    # Right-side axis showing empty cell count per store
    ax_right = ax.twinx()
    ax_right.set_ylim(ax.get_ylim())
    ax_right.set_yticks(np.arange(len(selected_stores)))
    right_labels = [str(c) if c > 0 else '' for c in empty_cells_counts]
    ax_right.set_yticklabels(right_labels, fontsize=9)
    for tick, cnt in zip(ax_right.get_yticklabels(), empty_cells_counts):
        if cnt > 0:
            tick.set_color('red')
            tick.set_fontweight('bold')
    ax_right.set_ylabel('Empty SKU Cells', fontsize=12)
    ax_right.tick_params(axis='y', direction='in')
    
    # Stats box
    total_allocated = matrix_data.sum()
    filled_combinations = np.count_nonzero(matrix_data)
    stats_text = (
        f"Total Allocated: {total_allocated:,}\\n"
        f"Filled Cells: {filled_combinations}\\n"
        f"Avg Empty Cells/store: {avg_empty_cells:.1f}\\n"
        f"Avg Color Coverage: {avg_color_cov:.2f}\\n"
        f"Avg Size Coverage: {avg_size_cov:.2f}"
    )
    
    fig.text(0.98, 0.98, stats_text, transform=fig.transFigure,
             fontsize=11, ha='right', va='top',
             bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.85))
    
    ax.set_title(f'{prod_cd} Allocation Matrix\\n({len(selected_stores)} Stores × {len(selected_skus)} SKUs)', 
                 fontsize=14, fontweight='bold', pad=20)
    ax.set_xlabel('SKU (Color-Size)', fontsize=12)
    ax.set_ylabel('Store ID (QTY_SUM)', fontsize=12)
    
    plt.tight_layout()
    
    # 저장
    save_path = os.path.join(output_dir, f"{prod_cd}_allocation_matrix.png")
    plt.savefig(save_path, dpi=300, bbox_inches='tight')
    print(f"   📊 배분 매트릭스 저장: {save_path}")
    plt.close()
    
    print(f"   📋 매트릭스 요약:")
    print(f"      표시된 매장: {len(selected_stores)}개")
    print(f"      표시된 SKU: {len(selected_skus)}개")
    print(f"      총 배분량: {total_allocated:,}개")
    print(f"      평균 빈 셀 수: {avg_empty_cells:.1f}")
    print(f"      평균 색상 커버리지: {avg_color_cov:.2f}")
    print(f"      평균 사이즈 커버리지: {avg_size_cov:.2f}")
    
    return {
        'prod_cd': prod_cd,
        'selected_stores': selected_stores,
        'selected_skus': selected_skus,
        'matrix_data': matrix_data,
        'total_allocated': total_allocated,
        'avg_empty_cells': avg_empty_cells,
        'avg_color_cov': avg_color_cov,
        'avg_size_cov': avg_size_cov
    }

# 4. 각 PROD_CD별로 시각화 실행
prod_codes = df['PROD_CD'].unique()
results = {}

for prod_cd in prod_codes:
    result = create_allocation_matrix_heatmap_by_prod(df, prod_cd, output_dir)
    if result:
        results[prod_cd] = result

print(f"\n🎯 총 {len(results)}개의 PROD_CD 시각화 완료!")
print("📁 저장 위치:", os.path.abspath(output_dir))
```
